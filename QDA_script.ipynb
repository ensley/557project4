```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn
from QDA import LDA,PCA,cross_validation
from QDA.LDA import DA
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data',
                  names= ['id','thickness','size.unif','shape.unif','adhesion',
                          'size.cell','nuclei','chromatin','nucleoli','mitoses','label'], 
                  na_values="?")
data = data.dropna()

train = data[:410]
test = data[410:]
```

```python
train.ix[:,1:].hist(figsize=(15,15), bins=20)
```

```python
model = DA(train.values[:, 1:-1], train.values[:,-1], mode='QDA')
model.sampleEstimate()
model.fit()
```

```python
y_hat, error = model.predict(test.values[:,1:-1], test.values[:,-1])
```

```python
error
```

```python
X = data.values[:,1:-1]
X = (X - X.mean()) / (X.max() - X.min())
```

```python
pca = PCA.PCA(X)
plt.bar(range(9),pca.summary(9))
sum(pca.summary(2))
```

```python
x = pca.transform(2)

model2 = DA(x[:410], data.values[:410,-1], mode="QDA")
model2.fit()
pred2 , err2 = model2.predict(x[410:],data.values[410:,-1])
err2 
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,1,1)
plt.scatter(x[410:,0], x[410:,1], s=50, alpha=.6, c=data.label.values[410:] , cmap=plt.cm.rainbow)
plt.title("True")
plt.subplot(2,1,2)
plt.scatter(x[410:,0], x[410:,1],  s=50, alpha=.6, c=pred2, cmap=plt.cm.rainbow)
plt.title("Predicted")
```
